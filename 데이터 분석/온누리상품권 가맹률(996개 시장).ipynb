```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('소상공인시장진흥공단_전국 온누리상품권 가맹점 현황_20221013.csv', encoding='cp949')
data
```

```python
data['소속 시장명(또는 상점가)'].value_counts()
```

```python
data['가맹점점포수'] = data.groupby('소속 시장명(또는 상점가)')['가맹점명'].transform('count')
data
```

```python
df = data.groupby('소속 시장명(또는 상점가)', as_index=False)[['가맹점명']].count()
df.rename(columns={'소속 시장명(또는 상점가)' : '시장명',
                  '가맹점명' : '가맹점수'}, inplace=True)
df
```

```python
df2 = pd.read_csv('전국전통시장표준데이터.csv', encoding='cp949')
df2 = df2[['시장명', '시장유형', '위도', '경도', '점포수', '사용가능상품권', '공중화장실보유여부', '주차장보유여부']]
df2
```

```python
merge_data = pd.merge(df, df2, how='inner', on='시장명')
merge_data
```

```python
merge_data['가맹률'] = merge_data['가맹점수'] / merge_data['점포수'] * 100
merge_data
```

**시장별 온누리상품권 가맹점수(온누리상품권 가맹점 정보) / 시장별 점포수(전국전통시장표준 데이터**

-> 전국전통시장표준 데이터에 나와있는 시장 수(1521개)와 온누리상품권 가맹점 데이터의 시장 수(1814) 정보가 많이 겹치지 않아서 가맹률을 구할 수 있는 시장 수가 996개로 줄어듦

-> 점포수보다 가맹점수가 더 많은 경우가 있어 신뢰도 급 하락
